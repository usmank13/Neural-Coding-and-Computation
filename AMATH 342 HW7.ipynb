```python
# Import statements

import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt

#Task 1, plotting exp(sin(t))

# generate time vector
t = np.arange(0,1.01, 0.01)

# generate function values
f = np.exp(np.sin(t))

# plot
plt.figure()
plt.plot(t,f)
plt.show()
plt.title('y = exp(sin(t))')
plt.xlim(0,1)
plt.ylabel('y')
plt.xlabel('t')
```

```python
# Task 3, solvin RC circuit equation and plotting voltage vs time

import scipy.integrate as si

v0 = 1
t1 = np.arange(0, 2, 0.01)

#dv_dt = -v_t / (R*C) 

# this defines the derivative
def dv_dt(v, t):
    #what are these RC constants????
    R = 1
    C = 0.1
    return -1*v / (R*C)

v_t = si.odeint(dv_dt, v0, t1)

plt.figure()
plt.plot(t1, v_t)
plt.title("Voltage vs. Time")
plt.xlim(0,2)
```

The graph for the solution V(t) yields an exponential curve with decay constant RC, which is what we expect of the solution.
